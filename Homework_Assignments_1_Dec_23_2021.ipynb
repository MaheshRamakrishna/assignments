```python
#11. write a function which can read video file and play for you. hint: explore
import cv2

def read_play_video(filepath, mirror=False):
    """
    Reads a local video and plays it.
    """
    capture = cv2.VideoCapture(filepath)
    cv2.namedWindow('Video', cv2.WINDOW_AUTOSIZE)

    while True:
        ret_val, frame = capture.read()

        if mirror:
            frame = cv2.flip(frame, 1)

        cv2.imshow('Video', frame)

        if cv2.waitKey(1) == 27:
            break


    cv2.destoryAllWindows()
    
read_play_video('/Users/mahesh.ramakrishna/Classes/resource/simple.mp4', mirror=True)
```

```python
#20. write a function by which you will be able to append two PDF files.

import PyPDF2

def merge_pdf_file(filePath1, filePath2, outputFilePath):
    """
    Merges two pdf files.
    """
    pdfFile1 = open(filePath1,'rb')
    pdfFile2 = open(filePath2,'rb')

    pdfReader1 = PyPDF2.PdfFileReader(pdfFile1)
    pdfReader2 = PyPDF2.PdfFileReader(pdfFile2)

    pdfWriter = PyPDF2.PdfFileWriter()

    for page_num in range(pdfReader1.numPages):
        pageObj = pdfReader1.getPage(page_num)
        pdfWriter.addPage(pageObj)

    for page_num in range(pdfReader2.numPages):
        pageObj = pdfReader2.getPage(page_num)
        pdfWriter.addPage(pageObj)

    pdfOutputFile = open(outputFilePath,'wb')
    pdfWriter.write(pdfOutputFile)

    pdfOutputFile.close()
    pdfFile1.close()
    pdfFile2.close()
    

filePath1 = "/Users/mahesh.ramakrishna/Classes/resource/simple1.pdf"
filePath2 = "/Users/mahesh.ramakrishna/Classes/resource/simple2.pdf"
outputFilePath = "/Users/mahesh.ramakrishna/Classes/resource/MergedFile.pdf"

merge_pdf_file(filePath1, filePath2, outputFilePath)
```
