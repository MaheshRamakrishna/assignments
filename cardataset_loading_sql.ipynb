```python
import mysql.connector as conn

mydb = conn.connect(host = "localhost", user = "root", passwd = "***")
```

```python
cursor = mydb.cursor()
```

```python
"""
q2 . create a database called as cardataset

"""
cursor.execute("CREATE DATABASE cardataset")
```

```python
cursor.execute("USE cardataset")
```

```python
"""
q3 . create a table called as car with a column name given in dataset description

"""
cursor.execute("CREATE TABLE cardataset.car(buying VARCHAR(30), maint VARCHAR(30), doors VARCHAR(30), persons VARCHAR(30), lugboot VARCHAR(30), safety VARCHAR(30), class VARCHAR(30))")
mydb.commit()
```

```python
cursor.execute("CREATE TABLE cardataset.car(buying VARCHAR(30), maint VARCHAR(30), doors VARCHAR(30), persons VARCHAR(30), lug_boot VARCHAR(30), safety VARCHAR(30), class VARCHAR(30))")
mydb.commit()
```

```python
"""
q4 . Dump all the data into car table 

"""
import csv
with open("/Users/mahesh.ramakrishna/Classes/resource/car.data", "r") as f:
    data = csv.reader(f, delimiter = '\n')
    for i in data:
        l = str(i[0]).split(",")
        s = "'"
        s += "', '".join(l)
        s += "'"
        cursor.execute(f"INSERT INTO cardataset.car VALUES({s})")
mydb.commit()
```

```python
cursor.execute("SELECT * FROM cardataset.car")
cursor.fetchall()
```

```python
"""
q5 . try to check whttps://archive.ics.uci.edu/ml/datasets/Car+Evaluationeather all the data is aviable inside your table or not 

"""
cursor.execute("SELECT count(*) FROM cardataset.car")
cursor.fetchall()
```

```python
"""
q6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value 

"""
cursor.execute("SELECT GROUP_CONCAT(maint, doors, persons, lug_boot, safety) grouped, count(*) FROM cardataset.car GROUP BY buying")
cursor.fetchall()
```

```python
"""
q7 . Try to filter a record where col3 value will be 4 .

"""

cursor.execute("SELECT * FROM cardataset.car WHERE doors = '4'")
cursor.fetchall()
```

```python
"""
q8.  Try to update a col 3 value with 8 whereever you have value equal to 2 

"""

cursor.execute("SELECT count(*) FROM cardataset.car WHERE doors = '2'")
cursor.fetchall()
# 432 count with 2 doors column



```

```python
cursor.execute("UPDATE cardataset.car SET doors = '8' WHERE doors = '2'")

cursor.execute("SELECT count(*) FROM cardataset.car WHERE doors = '2'")
cursor.fetchall()
# 0 count with 2 doors column
```
